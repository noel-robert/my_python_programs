```python
''' 
Create a class Vehicle.
create multiple attributes and methods for it
'''

import time


class Vehicle:
    def __init__(self, generalType):
        self.generalType = generalType
        self.fuelLevel = 100
        self.odometerReading = 0
        self.isRunning = False
        
        
    def inputDetails(self, name, model, fuelType, topSpeed, acceleration, horsepower, torque, maxRange, wheelbase, turningRadius, loadCapacity, axles, color):
        self.name = name
        self.model = model
        self.fuelType = fuelType
        self.topSpeed = topSpeed
        self.acceleration = acceleration
        self.horsepower = horsepower
        self.torque = torque
        self.wheelbase = wheelbase
        self.turningRadius = turningRadius
        self.loadCapacity = loadCapacity
        self.axles = axles
        self.maxRange = maxRange
        self.color = color


    def showDetails(self):
        print("Basic Vehicle Details - ")
        print(f"Name: {self.name} \nModel: {self.model}\n")
        print("Vehicle Technical Specs - ")
        print(f"Acceleration: {self.acceleration}m\s\s \nTop Speed: {self.topSpeed}kmph \nTorque: {self.torque}Nm\n")
        print("Other specs - ")
        print(f"Color: {self.color} \nRange: {self.maxRange}km \nTurning Radius: {self.turningRadius}m \n")


    def moveVehicle(self):
        if self.fuelLevel <= 0:
            print("Time to Refuel")
            return
        if not self.isRunning:
            print("Start the vehicle first!!\n")
            return

        self.currentSpeed = 0

        for i in range(10):
            print("."*i, end='')
            print("C", end='')
            print("."*(10-1-i))
            time.sleep(1)

        print()


    def startVehicle(self):
        if self.isRunning:
            print("Vehicle already started!\n")
        else:
            self.isRunning = True
            print("Vehicle started!\n")

    def stopVehicle(self):
        if self.isRunning:
            self.isRunning = False
            print("Vehicle stopped!\n")
        else:
            print("Vehicle is already stopped")



v1 = Vehicle("Car")
v1.inputDetails("Concept Car", "Mark1", "Electric", 150, 5, 400, 100, 500, 1240, 5, 0, 2, "Black")

v1.startVehicle()
v1.moveVehicle()
v1.stopVehicle()

v1.showDetails()
```
