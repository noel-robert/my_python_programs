```python
import threading
import time

def f(i):
    time.sleep(i)
    return

# threads
t1 = threading.Thread(target=f, args=(1.2,), name="Thread#1")
t1.start()

t2 = threading.Thread(target=f, args=(2.2,), name="Thread#2")
t2.start()

for p in range(5):
    time.sleep(p * 0.5)
    print('[',time.ctime(),']', t1.getName(), t1.is_alive())
    print('[',time.ctime(),']', t2.getName(), t2.is_alive())
    print()
```

```python
import threading
import time

def f(i):
    for p in range(3):
        time.sleep(i+1.5)
        print(threading.current_thread().getName())
    return

for i in range(3):
    t = threading.Thread(target=f, args=(i,))
    t.setName("Thread#"+str(i))
    t.start()
```

```python
'''
create 2 functons:
1. print alphabets from A - Z which is in a list
2. print numbers from 1 - 7

both uses for loop
'''


import time
import threading

def printAlphabets(i):
    for alphabet in range(65, 90+1):
        time.sleep(1)
        print(chr(alphabet))
        
def printNumbers(i):
    for number in range(0, 7+1):
        time.sleep(1)
        print(number)
        
        
alphabetThread = threading.Thread(target=printAlphabets, args=(1,), name="Thread_Alphabets")
numbersThread = threading.Thread(target=printNumbers, args=(2,), name="Thread_Numbers")
alphabetThread.start()
numbersThread.start()
```

```python

'''
question, challenge:
Producer Consumer problem

common thing is a queue
producer adds data
consumer removes from queue
implement using concept of threads

program stops when queue is full and producer cant add anymore
'''


import time
import threading
from collections import deque
import random

# maximum allowed size for the queue
max_size = 10


def producer(queue):
    '''
    add items to the queue if there is space
    '''
    while True:
        if len(queue) < max_size:
            queue.append("item")
            print("Item added to queue")
            time.sleep(random.uniform(0.1, 0.5))
        else:
            print("Queue is full")
            time.sleep(3)


def consumer(queue):
    '''
    consume items in queue if present
    '''
    while True:
        if len(queue) == 0:
            print("Empty queue, cannot remove items")
            time.sleep(5)
        else:
            print("Removing item from queue")
            queue.pop()
            time.sleep(random.uniform(0.3, 0.7))




queue = deque()
producerThread = threading.Thread(target=producer, args=(queue,))
consumerThread = threading.Thread(target=consumer, args=(queue,))
producerThread.start()
consumerThread.start()

# print("Program Execution Completed")
producerThread.join()
consumerThread.join()
```
